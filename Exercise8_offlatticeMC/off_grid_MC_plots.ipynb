```python
import numpy as np 
import matplotlib.pyplot as plt
import csv
import pandas as pd

d=[0.01,0.3,0.5,0.7,1.]
energies = []
with open('energy_hard_spheres.dat') as csvfile:
    contents= csv.reader(csvfile, delimiter=",")
    #next(contents)#skip header line
    for row in contents: 

        energies.append(np.asarray(row)[:-1:].astype('double'))

print(np.mean(energies[0:10],axis=1) )   
print(np.mean(energies[10:20]) ) 
print(np.mean(energies[20:30]) )     
print(np.mean(energies[30:40]) ) 
print(np.mean(energies[40:50]) )   

print(np.shape((energies)))
fig,axs=plt.subplots(4,5,figsize= (25,30))
axs=axs.flatten()
for ind,j in enumerate (np.arange(0,200,10)):
    for i in range(0,10):
        axs[ind].set_yscale("log")
        axs[ind].plot(energies[j+i])
 






```
