```python
import numpy as np 
import matplotlib.pyplot as plt
import csv
import pandas as pd

d=[0.01,0.3,0.5,0.7,1.]
energies = []
with open('energy_hard_spheres.dat') as csvfile:
    contents= csv.reader(csvfile, delimiter=",")
    #next(contents)#skip header line
    for row in contents: 

        energies.append(np.asarray(row)[:-1:].astype('double'))
print(np.mean(energies[0:10]) )   
print(np.mean(energies[10:20]) ) 
print(np.mean(energies[20:30]) )     
print(np.mean(energies[30:40]) ) 
print(np.mean(energies[40:]) )   
mean_energies=np.mean(energies,axis=1).reshape(10,5)


print(np.mean(mean_energies,axis=0))

plt.plot(d,np.mean(mean_energies,axis=0))




```
